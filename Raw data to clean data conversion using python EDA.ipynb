# Missing value Treatment for numerical data

```python
clean_data.isnull().sum()
```

```python
clean_data['Age']
```

```python
import numpy as np #import numpy
```

```python
#For numerical value  we use mean, median and mode strategy
clean_data['Age']= clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age']))) #using mean strategy
```

```python
clean_data['Age'] #Displayes the cleane Age column
```

```python
clean_data['Exp']=clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp']))) #here also goes the same
```

```python
clean_data['Exp'] #prints it
```

```python
clean_data['Location'].isnull().sum() #here we check rather there is any missing value or not
```

```python
clean_data['Location'] #displayes the location column
```

```python
#for categorical value we use mode strategy or KNN strategy
clean_data['Location'] = clean_data['Location'].fillna(clean_data['Location'].mode()[0]) #using mode strategy
```

```python
clean_data['Location'] # prints the location column
```

```python
clean_data #displayes the clean_data
```

```python
clean_data.info() #see the informations
```

```python
#For change the type -- numerical
clean_data['Age'] = clean_data['Age'].astype(int) #Chenge flote or string type to integer type
```

```python
clean_data['Age'] 
```

```python
clean_data['Salary']=clean_data['Salary'].astype(int) #change the type
```

```python
clean_data['Salary']
```

```python
clean_data['Exp']=clean_data['Exp'].astype(int) #change the type
```

```python
clean_data['Exp']
```

```python
# change the type -- categorical
clean_data['Name'] = clean_data['Name'].astype('category')
clean_data['Domain'] = clean_data['Name'].astype('category')
clean_data['Location'] = clean_data['Name'].astype('category')
```

```python
clean_data.info() #see the information
```
